```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
catalog_df = pd.read_csv('../data/processed/catalog.csv')
returns_df = pd.read_csv('../data/processed/returns.csv')
print("Datasets loaded.")

# Visualize the most common product types
plt.figure(figsize=(12, 6))
sns.countplot(y='article_type', data=catalog_df, order=catalog_df['article_type'].value_counts().head(10).index)
plt.title('Top 10 Most Common Product Types in Catalog')
plt.show()

# Visualize the most returned product classes
plt.figure(figsize=(12, 6))
sns.countplot(y='class_name', data=returns_df, order=returns_df['class_name'].value_counts().head(10).index)
plt.title('Top 10 Most Returned Product Classes')
plt.show()
```

```python
print("--- Catalog Data Info ---")
catalog_df.info()

print("\n--- Top 10 Most Common Product Types ---")
print(catalog_df['article_type'].value_counts().head(10))
```

```python
# Plot the top 10 most common product types
plt.figure(figsize=(12, 6))
sns.countplot(y='article_type', data=catalog_df, order=catalog_df['article_type'].value_counts().head(10).index)
plt.title('Top 10 Most Common Product Types in Catalog')
plt.xlabel('Count')
plt.ylabel('Product Type')
plt.show()
```

```python
print("--- Returns Data Info ---")
returns_df.info()

print("\n--- Top 10 Most Returned Product Classes ---")
print(returns_df['class_name'].value_counts().head(10))
```

```python
# Plot the most frequently returned product classes
plt.figure(figsize=(12, 6))
sns.countplot(y='class_name', data=returns_df, order=returns_df['class_name'].value_counts().head(10).index)
plt.title('Top 10 Most Returned Product Classes')
plt.xlabel('Number of Returns')
plt.ylabel('Product Class')
plt.show()
```
